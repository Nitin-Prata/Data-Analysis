```python
final = pd.DataFrame()

# The maximum allowed page is 500
for j in range(1, 501):
    url = f'https://api.themoviedb.org/3/movie/popular?language=en-US&page={j}&api_key=51566c3235153d1ad6b26394aaa1791a'
    response = requests.get(url).json()
    
    # Check if 'results' key exists in response
    if 'results' not in response:
        print(f"Page {j}: {response.get('status_message', 'Unknown error')}")
        continue

    popularity = []
    vote_average = []
    title = []
    release_date = []
    original_title = []
    vote_count = []
    overview = []
    original_language = []

    for i in response['results']:
        popularity.append(i['popularity'])
        vote_average.append(i['vote_average'])
        title.append(i['title'])
        release_date.append(i['release_date'])
        original_title.append(i['original_title'])
        vote_count.append(i['vote_count'])
        overview.append(i['overview'])
        original_language.append(i['original_language'])

    d = {
        'title': title,
        'overview': overview,
        'original_language': original_language,
        'release_date': release_date,
        'popularity': popularity,
        'vote_count': vote_count,
        'vote_average': vote_average
    }
    df = pd.DataFrame(d)
    final = pd.concat([final, df], ignore_index=True)

```
